# PYTHON WEB APPLICATION

```python
import pandas as pd
import json
from flask import Flask, render_template
```

## 1) Read CSV and drop records without company name

```python
df = pd.read_csv("/Users/sarahdoctor/Desktop/DataManagement/crunchbase_orgs.csv")
```

```python
df.dropna(subset=['name'], inplace=True)
```

#### Number of records

```python
print(f"There are {len(df)} records.")
```

#### First 10 Records

```python
df.head(10)
```

## 2) USA-based companies whose name starts with "ad" 

```python
USData = df[(df['country_code'] == 'USA') & (df['name'].str.lower().str.startswith('ad'))]
```

#### Number of Records

```python
print(f"US-based companies starting with 'ad': {len(USData)}")
```

#### First 10 Records

```python
USData.head(10)
```

## 3) Data Frame as a list of JSON objects

```python
json_list = USData.to_json(orient='records')
```

```python
# Parse the JSON string back into a list to count the JSON objects
json_objects = json.loads(json_list)
```

#### Count of JSON Objects

```python
print(f"Count of JSON objects: {len(json_objects)}")
```

#### First 5 JSON objects

```python
for json_object in json_objects[:5]:
    print(json.dumps(json_object, indent=2))
```

#### JSON objects as a text (JSON string) file

```python
with open('data.json', 'w') as file:
    file.write(json_list)
```

#### Number of records in file

```python
with open('data.json', 'r') as file:
    file_content = file.read()
    records = json.loads(file_content)
    print(f"Number of records in the resulting file: {len(records)}")
```

## 4) Read objects back into a dataframe keeping only NY-Based companies

```python
newDf = pd.read_json('data1.json')
newDf = newDf[newDf['region'] == 'New York']
```

#### Number of results

```python
print(f"Number of records in New York: {len(newDf)}")
```

#### Output as a Flask webpage

```python
app = Flask("CrunchBaseData")

@app.route('/')
def show_data():
    return newDf.to_html()

app.run(host='localhost', port=4200)
```
